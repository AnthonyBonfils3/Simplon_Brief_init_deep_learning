```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
import math

print ('TensorFlow version: ' + tf.__version__)

# Create noisy data
x_data = np.linspace(-10, 10, num=1000)
y_data = 0.1*x_data*np.cos(x_data) + 0.1*np.random.normal(size=1000)
print('Data created successfully')
```

## display curve to fit

```python
# Display the dataset
plt.scatter(x_data[::1], y_data[::1], s=2)
plt.grid()
plt.show()
plt.savefig('./figures/dataset.png',dpi=300)
f = open('./figures/dataset.png', "r")
```

## Create the model

Create the model with two 64 units hidden layers

```python
# Create the model 
def create_model(n_hyden_layer=2, n_perceptron=64, activate_fcn='relu'):
    model = keras.Sequential()
    model.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
    for i in range(n_hyden_layer):
        model.add(keras.layers.Dense(units = n_perceptron, activation = activate_fcn))
    model.add(keras.layers.Dense(units = 1, activation = 'linear'))
    model.compile(loss='mse', optimizer="adam")
    return model

##n_hyden_layer=2
##n_perceptron = 64
#n_hyden_layer=1
#n_perceptron = 4096
n_hyden_layer = 4
n_perceptron = 32
activate_fcn = 'elu'
model = create_model(n_hyden_layer, n_perceptron, activate_fcn)
# Display the model
model.summary()
```

## Training over 100 epochs

```python
# Training
model_conv = model.fit( x_data, y_data, epochs=500, verbose=1) ## de base 100 epochs
```

```python
# list all data in history
print(model_conv.history.keys())
# summarize history for accuracy
plt.plot(model_conv.history['loss'])
plt.ylabel('loss')
plt.xlabel('epochs')
plt.title(f'MSE = f(epochs) model Multiple Layers Perceptron (Layers: {n_hyden_layer+2}, n_perseptron : {n_perceptron}')
```

## Predict and display

```python
# Compute the output 
y_predicted = model.predict(x_data)

# Display the result
plt.scatter(x_data[::1], y_data[::1], s=1)
plt.plot(x_data, y_predicted, 'r', linewidth=4)
plt.grid()
#plt.show()
plt.savefig('./figures/training.png', dpi=300)
f = open('./figures/training.png', "r")
```

```python
from mpl_toolkits.mplot3d import Axes3D
def draw3Dim (x, y, z, z_pred=[]):
    fig = plt.figure(figsize=[10,10])
    ax = fig.add_subplot(111, projection='3d')

    ax.scatter(x, y, z, marker='o')
    
    try:
        ax.plot3D(x, y, z_pred, 'r', linewidth=4)
    except:
        pass
        
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.title("Repr√©sentation du dataset r√©duit en 3 dimensions ")

    plt.show()
    
draw3Dim(x_data, y_data, z_data)
```
